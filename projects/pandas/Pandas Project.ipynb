## Defining some helper functions

```python
def year_prettify(year):
    if year > 1000: 
        return year
    else:
        return np.nan
    
def sex_prettify(sex):
    if sex == 'M' or sex == 'F':
        return sex
    else:
        return np.nan
    
def age_prettify(age):
    try:
        age = int(age)
    except ValueError:
        age = 0
        
    if (age > 0 and age <= 100):
        return age
    else:
        return np.nan
    
def fatal_prettify(fatal):
    if fatal == 'N' or fatal == 'Y':
        return fatal
    else:
        return np.nan
    
def date_prettify(date):
    months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    num_months = ['01', '02', '03', '04', '05', '06', '07', '08', '09', '10', '11', '12']
    
    date = str(date)
    if (len(date) == 11 and date[2] == '-'):
        for i in range(len(months)):
            date = date.replace(months[i], num_months[i])
        return date
    else:
        return np.nan
```

## Data cleaning

Fixing columns names

```python
names = list(df.columns)
names[9] = 'Sex'
names[12] = 'Fatal'
names[14] = 'Species'
df.columns = names
names
```

Prettyfing values

```python
df['Year'] = df['Year'].apply(year_prettify)
df['Sex'] = df['Sex'].apply(sex_prettify)
df['Age'].fillna(0, inplace=True)
df['Age'] = df['Age'].apply(age_prettify)
df['Fatal'] = df['Fatal'].apply(fatal_prettify)
df['Date'] = df['Date'].apply(date_prettify)
```
